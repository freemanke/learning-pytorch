# 绘图
[参考链接](https://www.runoob.com/matplotlib/matplotlib-tutorial.html)

```python
from matplotlib import pyplot as plt
import numpy as np

x = np.arange(21)
y = x**2 + 2

plt.xlabel("X")
plt.ylabel("Y")
plt.title("y=x^2+2")
plt.plot(x,y)
plt.grid()
plt.show()
```

## 饼图

```python
x = np.array([20,30,40,10])
plt.pie(x,
        labels=['A','B','C','D'],
        explode=(0,0.2,0,0),
        colors=['green','yellow','red','blue'],
        autopct="%.2f%%"
        )
plt.title("PIE of PERCENTAGE")
plt.show()
```

## 散点图

```python
x = np.arange(-10,11)
y = x**3

plt.xlabel("X")
plt.ylabel("Y")
plt.title("Scatter")
plt.scatter(x,y)
plt.grid()
plt.show()
```

## 柱状图

```python
x = np.array(['Jen','Feb','Mar','April','May'])
y = np.array([20,100,30,45,30])
plt.title('Sales')
plt.bar(x,y)
plt.show()
```
