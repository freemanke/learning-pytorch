# 图像处理

```python
from PIL import Image
import numpy as np
from matplotlib import pyplot as plt

image = Image.open("./data/kekexin.jpg")
print(image.size)

image_pillow = np.asarray(image)
print(image_pillow.shape)

plt.subplot(2,2,1)
plt.imshow(image_pillow)

plt.subplot(2,2,2)
plt.imshow(image_pillow)

plt.subplot(2,2,3)
plt.imshow(image_pillow)

plt.subplot(2,2,4)
plt.imshow(image_pillow)
plt.show()
```
